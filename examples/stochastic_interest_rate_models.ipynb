```python
from quantutils.models.interest_rates import merton, vasicek
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("seaborn")
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
n_steps = 10000
t, xt = merton(mu = 0.05,
               sigma = 0.3,
               x0 = 0.01,
               n_steps= n_steps,
               dt= 1.0/n_steps)
plt.plot(t, xt)
plt.xlabel("t")
plt.ylabel("IR")
_ = plt.title("Interest Rate simulation - Merton model ")
```

```python
t, xt = vasicek(a = 0.7,
                b = 1.0,
                sigma = 0.04,
                x0 = 0.3,
                n_steps=20000,
                dt=0.001)
plt.plot(t, xt)
plt.xlabel("t")
plt.ylabel("IR")
_ = plt.title("Interest Rate simulation - Vasicek model")
```
